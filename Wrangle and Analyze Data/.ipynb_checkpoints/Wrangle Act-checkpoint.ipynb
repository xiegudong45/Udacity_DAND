```python
my_demo_lst = []
with open('tweet_json.txt', encoding='utf-8') as json_file:  
    all_data = json.load(json_file)
    for dictionary in all_data:
        tw_id = dictionary['id']
        tweet_text = dictionary['text']
        origin_url = tweet_text[tweet_text.find('https'):]
        favorite_count = dictionary['favorite_count']
        retweet_count = dictionary['retweet_count']
        followers_count = dictionary['user']['followers_count']
        friends_count = dictionary['user']['friends_count']
        whole_source = dictionary['source']
        only_device = whole_source[whole_source.find('rel="nofollow">') + 15:-4]
        source = only_device
        retweeted_status = dictionary['retweeted_status'] = dictionary.get('retweeted_status', 'Original tweet')
        if retweeted_status == 'Original tweet':
            url = origin_url
        else:
            retweeted_status = 'This is a retweet'
            url = 'Retweet'

        my_demo_lst.append({'tw_id': tw_id,
                             'favorite_count': int(favorite_count),
                             'retweet_count': int(retweet_count),
                             'followers_count': int(followers_count),
                             'friends_count': int(friends_count),
                             'url': url,
                             'source': source,
                             'retweeted_status': retweeted_status,
                            })
        df_tweet_json = pd.DataFrame(my_demo_lst, columns = ['tw_id', 'favorite_count','retweet_count', 
                                                           'followers_count', 'friends_count','source', 
                                                           'retweeted_status', 'url'])
```

```python
def get_decimal_rating(df):
    pattern = '(\d+\.\d+)\/(\d+)'
    matches = re.search(pattern, df['text'])
    if matches != None:
        df['rating_numerator'] = matches.group(1)
        df['rating_denominator'] = matches.group(2)
    return df
df_twi_clean = df_twi_clean.apply(get_decimal_rating, axis=1)
```

#### 8.
* `doggo`, `floofer`, `pupper`, `puppo` can be combined into one column
These data frames can be merged into one.

##### Define
Make one column for dog stage (doggo, floofer, pupper, and puppo) by saving value ('None' if no dog stage given). Also record if there are multiple dog stages, separating by a comma.

#### Code

```python
df_twi_clean['dog_stage'] = 'None'

def get_dog_stage(df):
    dog_stage = []
    if df['doggo'] == 'doggo':
        dog_stage.append('doggo')
    if df['floofer'] == 'floofer':
        dog_stage.append('floofer')
    if df['pupper'] == 'pupper':
        dog_stage.append('pupper')
    if df['puppo'] == 'puppo':
        dog_stage.append('puppo')
    if len (dog_stage) >= 1:
        df['dog_stage'] = ','.join(dog_stage)
    else:
        df['dog_stage'] = 'None'
    return df
df_twi_clean = df_twi_clean.apply(get_dog_stage, axis=1)
# df_twi_clean.iloc[29]
df_twi_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

### 1. Find the relationship between favorate_count and retweet_count

```python
from scipy import stats
```

```python
df_1 = df[['favorite_count', 'retweet_count', 'dog_stage']]
df_1
```

```python
y = df_1.retweet_count
x = df_1.favorite_count
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept
plt.plot(x, y, 'o', x, line)
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.legend(['Raw Data', 'Regression Model'])
print('Linear Regressiong Model Equation: retweet_count = {} * favorite_count + ({})'.format(slope, intercept))
```

```python
print('Correlation Coefficient:', r_value)
```

I first created a scatter plot of all tweets and the ideal linear model. Note that the actual relationship may not be linear, but it can give us a general idea of the two columns. You can see that most of the tweets received a comparably low retweet and favorite count.

Since the correlation coefficient is approximately 0.928, it indicates that there's a strong positive relationship between favorite_count and retweet_count.

In this step, I will only choose the data with one dog stage and then get a linear model for each class.

```python
xlim = (0, 80000)
ylim = (0, 40000)

df_pupper = df[df['dog_stage'] == 'pupper']
df_puppo = df[df['dog_stage'] == 'puppo']
df_doggo = df[df['dog_stage'] == 'doggo']
df_floofer = df[df['dog_stage'] == 'floofer']

x_pupper = df_pupper.favorite_count
y_pupper = df_pupper.retweet_count
slope_pupper, intercept_pupper, r_value_pupper, p_value, std_err = stats.linregress(x_pupper, y_pupper)
line_pupper = slope_pupper * x_pupper + intercept_pupper
plt.plot(x_pupper, y_pupper, 'o', alpha=0.5)
plt.plot(x_pupper, line_pupper)
plt.xlim(xlim)
plt.ylim(ylim)
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Pupper')
plt.legend(['Raw Data', 'Regression Model'])
print('Linear Regressiong Model Equation: retweet_count = {} * favorite_count + ({})'.format(slope_pupper, intercept_pupper))
```

```python
x_puppo = df_puppo.favorite_count
y_puppo = df_puppo.retweet_count
slope_puppo, intercept_puppo, r_value_puppo, p_value, std_err = stats.linregress(x_puppo, y_puppo)
line_puppo = slope_puppo * x_puppo + intercept_puppo
plt.plot(x_puppo, y_puppo, 'o', alpha=0.5, color='Red')
plt.plot(x_puppo, line_puppo)
plt.xlim(xlim)
plt.ylim(ylim)
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Puppo')
plt.legend(['Raw Data', 'Regression Model'])
print('Linear Regressiong Model Equation: retweet_count = {} * favorite_count + ({})'.format(slope_puppo, intercept_puppo))
```

```python
x_doggo = df_doggo.favorite_count
y_doggo = df_doggo.retweet_count
slope_doggo, intercept_doggo, r_value_doggo, p_value, std_err = stats.linregress(x_doggo, y_doggo)
line_doggo = slope_doggo * x_doggo + intercept_doggo
plt.plot(x_doggo, y_doggo, 'o', alpha=0.5, color='Orange')
plt.plot(x_doggo, line_doggo)
plt.xlim(xlim)
plt.ylim(ylim)
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Doggo')
plt.legend(['Raw Data', 'Regression Model'])
print('Linear Regressiong Model Equation: retweet_count = {} * favorite_count + ({})'.format(slope_doggo, intercept_doggo))
```

```python
x_floofer = df_floofer.favorite_count
y_floofer = df_floofer.retweet_count
slope_floofer, intercept_floofer, r_value_floofer, p_value, std_err = stats.linregress(x_floofer, y_floofer)
line_floofer = slope_floofer * x_floofer + intercept_floofer
plt.plot(x_floofer, y_floofer, 'o', alpha=0.5, color='Green')
plt.plot(x_floofer, line_floofer)
plt.xlim(xlim)
plt.ylim(ylim)
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Floofer')
plt.legend(['Raw Data', 'Regression Model'])
print('Linear Regressiong Model Equation: retweet_count = {} * favorite_count + ({})'.format(slope_floofer, intercept_floofer))
```

From these four plots we can see that there're more tweets classified as pupper than any other classification, Doggo ranks the second, Puppo ranks the third, and Floofer has the least. Moreover, the retweet/favourite count ratio of Pupper tends to be the smallest among the four classes. Apart from pupper, the tweets which classified as the other three classes with lower value tend to have a higher ratio of retweets to favorites than the line of best fit.

```python
df_2 = df[['retweet_count', 'favorite_count', 'rating_numerator', 'rating_denominator', 'p1', 'p1_dog', 'p1_conf', 
           'p2', 'p2_dog', 'p2_conf', 'p3', 'p3_dog', 'p3_conf']]
df_2['rating'] = df_2['rating_numerator'] / df_2['rating_denominator']
df_2['breed'] = ''
df_2['breed_conf'] = 0
df_2.info()
```

```python
def breed_match(df):
    breed = 'unknown_breed'
    conf = 0
    predict = [(df[['p1', 'p1_dog', 'p1_conf']]), 
               (df[['p2', 'p2_dog', 'p2_conf']]),
               (df[['p3', 'p3_dog', 'p3_conf']])]
    idx_breed = 0
    idx_isDog = 1
    idx_conf = 2
    for pred in predict:
        if pred[0]:
            if pred[2] > conf:
                breed = pred[0]
                conf = pred[2]
    df['breed'] = breed
    df['breed_conf'] = conf
    return df
df_2 = df_2.apply(breed_match, axis=1)
df_2.drop(['p1', 'p1_dog', 'p1_conf', 'p2', 'p2_dog', 'p2_conf', 'p3', 'p3_dog', 'p3_conf'], axis=1, inplace=True)
```
